# PIPELINE

```python
from transformers import pipeline
```

```python
text_generator = pipeline("text-generation")
text_generator("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    ["This is a course about the Transformers library", "United States president is Biden"],
    candidate_labels=["education", "politics", "business"],
)
```

```python
generator = pipeline("text-generation")
text = generator("In this course, we will teach you how to",  max_length=30, num_return_sequences=2)
for t in text:
  print(t["generated_text"])
```

## Code Generation

```python
model = "Qwen/Qwen2.5-Coder-0.5B"
coder = pipeline("text-generation", model=model)
coder("Python code for prime number generator.", max_length=30)[0]["generated_text"]
```

```python
r1 = coder("Python code for prime number generator.", max_length=30)[0]["generated_text"]
print(r1)
```

## Name entity recognition (NER)

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

## Fill-mask

```python
unmasker = pipeline("fill-mask", model="bert-base-uncased")

result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

# TRANSFORMER

```python
from transformers import AutoTokenizer, AutoModel
```

```python
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModel.from_pretrained(checkpoint)

raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
    "I love chocolate!"
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
model.config.id2label
model.generate(outputs.logits)
```
